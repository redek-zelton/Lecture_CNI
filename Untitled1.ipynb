```python
def preProcessing(img,LongueurImage,HauteurImage):
    imgCopie=cv2.resize(img,(LongueurImage,HauteurImage))
    gray = cv2.cvtColor(imgCopie, cv2.COLOR_BGR2RGB)
    blured = cv2.GaussianBlur(gray,(5,5),cv2.BORDER_DEFAULT)
    imgCanny= cv2.Canny(blured, 20,10)
    kernel = np.ones((5,5))
    imgDial = cv2.dilate(imgCanny,kernel,iterations=2)
    imgThres = cv2.erode(imgDial,kernel,iterations=1)
    return imgThres
```

```python
# prendre le contour de la carte
def getContours(imgThres):
    biggest=np.array([])
    maxArea = 0
    contours,hierarchy = cv2.findContours(imgThres,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_NONE)
    for cnt in contours:
        area = cv2.contourArea(cnt)
        if area>500:
            cv2.drawContours(imgThres,cnt,-1,(120,120,120),3)
            peri = cv2.arcLength(cnt,True)
            approx = cv2.approxPolyDP(cnt,0.02*peri,True)
            if area > maxArea and len(approx) in (4,5,6,7,8,9,10):
                biggest = approx
                maxArea = area
    biggest=shape4(biggest)
    return biggest
```

```python
def shape4(biggest):
    biggest=biggest
    L1=[]
    L2=[]
    if biggest.shape != (4,1,2):
        for i in biggest:
            L1.append(i[0][0])
            L2.append(i[0][1])
        biggest=np.array([[[min(L1),min(L2)]],[[min(L1),max(L2)]],[[max(L1),min(L2)]],[[max(L1),max(L2)]]],np.int32)
    return biggest
```

```python
def reorder(myPoints):
    myPoints = myPoints.reshape((4,2))
    myPointNew=np.zeros((4,1,2),np.int32)
    add=myPoints.sum(1)
    #print("add",add)
    myPointNew[0]= myPoints[np.argmin(add)]
    myPointNew[3]= myPoints[np.argmax(add)]
    diff=np.diff(myPoints,axis=1)
    myPointNew[1]= myPoints[np.argmin(diff)]
    myPointNew[2]= myPoints[np.argmax(diff)]
    #print("NewPoints",myPointsNew)
    return myPointNew
```

```python
def resize(biggest):
    Long=[]
    Haut=[]
    for line in biggest:
        Long.append(line[0][0])
        Haut.append(line[0][1])
    return[max(Long)-min(Long), max(Haut)-min(Haut)]
```

```python
def getWarp(img,biggest):
    global LongueurImage
    global HauteurImage
    imgCopie=cv2.resize(img,(LongueurImage,HauteurImage))
    biggest = reorder(biggest)
    pts1=np.float32(biggest)
    pts2=np.float32([[0,0],[LongueurImage,0],[0,HauteurImage],[LongueurImage,HauteurImage]])
    #pts2=np.float32([[0,0],[HauteurImage,0],[0,LongueurImage],[HauteurImage,LongueurImage]])
    matrix = cv2.getPerspectiveTransform(pts1,pts2)
    #imgOutput = cv2.warpPerspective(imgCopie,matrix,(LongueurImage,HauteurImage))
    imgOutput = cv2.warpPerspective(imgCopie,matrix,(L1,H1))
    return imgOutput
```
